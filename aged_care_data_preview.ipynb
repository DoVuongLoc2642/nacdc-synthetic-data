# Aged Care Baseline Dataset Preview

This notebook provides a comprehensive preview and initial exploration of the aged care baseline synthetic dataset generated for research purposes.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Load the CSV Data
data_path = Path('data/agedcare_baseline_3000.csv')
print(f"Loading data from: {data_path}")

# Read the CSV file
df = pd.read_csv(data_path)
print(f"âœ… Successfully loaded aged care baseline dataset")
print(f"ðŸ“Š Dataset contains {len(df)} records")
```

```python
# Display Basic Dataset Information
print("=== DATASET OVERVIEW ===")
print(f"Shape: {df.shape}")
print(f"Columns: {len(df.columns)}")
print(f"Rows: {len(df)}")
print("\n=== COLUMN NAMES ===")
print(df.columns.tolist())
```

```python
# Preview Data Structure and Types
print("=== FIRST 5 ROWS ===")
display(df.head())

print("\n=== LAST 5 ROWS ===")
display(df.tail())

print("\n=== DATA TYPES ===")
print(df.dtypes)
```

```python
# Examine Missing Values
print("=== MISSING VALUES ANALYSIS ===")
missing_values = df.isnull().sum()
missing_percentages = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing Count': missing_values.values,
    'Missing %': missing_percentages.values
})

print("Missing values per column:")
display(missing_df[missing_df['Missing Count'] > 0])

if missing_df['Missing Count'].sum() == 0:
    print("âœ… No missing values found in the dataset!")
```

```python
# Generate Summary Statistics
print("=== NUMERICAL VARIABLES SUMMARY ===")
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numerical columns: {numerical_cols}")

if numerical_cols:
    display(df[numerical_cols].describe())
else:
    print("No numerical columns found")
```

```python
# Explore Categorical Variables
print("=== CATEGORICAL VARIABLES SUMMARY ===")
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns: {categorical_cols}")

for col in categorical_cols[:5]:  # Show first 5 categorical columns
    print(f"\n--- {col.upper()} ---")
    value_counts = df[col].value_counts()
    print(f"Unique values: {df[col].nunique()}")
    print("Top values:")
    display(value_counts.head(10))
```

```python
# Create Basic Visualizations - Demographics
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Aged Care Demographics Overview', fontsize=16, fontweight='bold')

# Age distribution
if 'age_at_admission' in df.columns:
    axes[0,0].hist(df['age_at_admission'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    axes[0,0].set_title('Age at Admission Distribution')
    axes[0,0].set_xlabel('Age')
    axes[0,0].set_ylabel('Frequency')

# Sex distribution
if 'sex' in df.columns:
    sex_counts = df['sex'].value_counts()
    axes[0,1].bar(sex_counts.index, sex_counts.values, color=['lightcoral', 'lightblue', 'lightgreen'])
    axes[0,1].set_title('Sex Distribution')
    axes[0,1].set_xlabel('Sex')
    axes[0,1].set_ylabel('Count')

# Care level distribution
if 'care_level' in df.columns:
    care_counts = df['care_level'].value_counts().sort_index()
    axes[1,0].bar(care_counts.index, care_counts.values, color='lightseagreen')
    axes[1,0].set_title('Care Level Distribution')
    axes[1,0].set_xlabel('Care Level')
    axes[1,0].set_ylabel('Count')

# Length of stay distribution
if 'length_of_stay_days' in df.columns:
    axes[1,1].hist(df['length_of_stay_days'], bins=30, alpha=0.7, color='orange', edgecolor='black')
    axes[1,1].set_title('Length of Stay Distribution')
    axes[1,1].set_xlabel('Days')
    axes[1,1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create Health and Care Visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Health and Care Characteristics', fontsize=16, fontweight='bold')

# Dementia status
if 'dementia_status' in df.columns:
    dementia_counts = df['dementia_status'].value_counts()
    axes[0,0].pie(dementia_counts.values, labels=dementia_counts.index, autopct='%1.1f%%', startangle=90)
    axes[0,0].set_title('Dementia Status Distribution')

# Medication count
if 'medication_count' in df.columns:
    axes[0,1].hist(df['medication_count'], bins=15, alpha=0.7, color='mediumpurple', edgecolor='black')
    axes[0,1].set_title('Medication Count Distribution')
    axes[0,1].set_xlabel('Number of Medications')
    axes[0,1].set_ylabel('Frequency')

# ACFI Care Domain scores
if 'acfi_care_domain' in df.columns:
    axes[1,0].hist(df['acfi_care_domain'], bins=20, alpha=0.7, color='gold', edgecolor='black')
    axes[1,0].set_title('ACFI Care Domain Scores')
    axes[1,0].set_xlabel('ACFI Score')
    axes[1,0].set_ylabel('Frequency')

# SEIFA decile
if 'seifa_decile' in df.columns:
    seifa_counts = df['seifa_decile'].value_counts().sort_index()
    axes[1,1].bar(seifa_counts.index, seifa_counts.values, color='lightsteelblue')
    axes[1,1].set_title('SEIFA Decile Distribution')
    axes[1,1].set_xlabel('SEIFA Decile')
    axes[1,1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Create Correlation Heatmap for Numerical Variables
if len(numerical_cols) > 1:
    plt.figure(figsize=(12, 8))
    correlation_matrix = df[numerical_cols].corr()
    
    # Create heatmap
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, 
                mask=mask,
                annot=True, 
                cmap='RdBu_r', 
                center=0,
                square=True,
                fmt='.2f',
                cbar_kws={"shrink": .8})
    
    plt.title('Correlation Matrix of Numerical Variables', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numerical variables for correlation analysis")
```

```python
# Dataset Quality Assessment
print("=== DATASET QUALITY ASSESSMENT ===")

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

# Check data ranges for key variables
if 'age_at_admission' in df.columns:
    age_stats = df['age_at_admission'].describe()
    print(f"\nAge statistics:")
    print(f"  Range: {age_stats['min']:.0f} - {age_stats['max']:.0f} years")
    print(f"  Mean: {age_stats['mean']:.1f} years")
    print(f"  Median: {age_stats['50%']:.1f} years")

if 'length_of_stay_days' in df.columns:
    los_stats = df['length_of_stay_days'].describe()
    print(f"\nLength of Stay statistics:")
    print(f"  Range: {los_stats['min']:.0f} - {los_stats['max']:.0f} days")
    print(f"  Mean: {los_stats['mean']:.1f} days")
    print(f"  Median: {los_stats['50%']:.1f} days")

# Check for logical consistency
if 'sex' in df.columns:
    sex_values = df['sex'].unique()
    print(f"\nSex categories: {sorted(sex_values)}")

if 'care_level' in df.columns:
    care_levels = df['care_level'].unique()
    print(f"Care levels: {sorted(care_levels)}")

print("\nâœ… Dataset preview completed successfully!")
```
